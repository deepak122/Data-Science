```python
##label encoding 

from sklearn import preprocessing
le=preprocessing.LabelEncoder()
for x in cat_df.columns:
      cat_df[x]=le.fit_transform(cat_df[x])   
      cat_df_test[x]=le.fit_transform(cat_df_test[x])
```

```python
#combining test and train dataset for one hot encoding 

train_len=len(train)
dataset = pd.concat(objs=[cat_df,cat_df_test], axis=0)
dataset.shape
```

```python
#one hot encodeing 

na=np.array(dataset)
# binary encode
onehot_encoder = preprocessing.OneHotEncoder()
onehot_encoded = onehot_encoder.fit_transform(na).toarray()


#onehot_encoded
temp_df=pd.DataFrame(data=onehot_encoded)
```

```python
#one hot encoded test and train datset 
df_ohe =temp_df[:train_len]
df_ohe_test = temp_df[train_len:]
```

```python
#radius skewness of train and test dataset before treatment
sns.kdeplot(num_df.radius)
sns.kdeplot(num_df_test.radius)
```

```python
# skewed dataset treatment by taking log 
#scaling dataset 

num_df.radius=preprocessing.scale(np.log(num_df.radius+1))
num_df_test.radius=preprocessing.scale(np.log(num_df_test.radius+1))
```

```python
#radius skewness of train dataset after treatment

sns.kdeplot(num_df.radius)
sns.kdeplot(num_df_test.radius)
```

```python
#weight before treatment train and test dataset

sns.kdeplot(num_df.weight)
sns.kdeplot(num_df_test.weight)
```

```python
#treatment of skewed datset by taking square root 
#scaling of dataset 

num_df.weight=preprocessing.scale(np.sqrt(num_df.weight))
num_df_test.weight=preprocessing.scale(np.sqrt(num_df_test.weight))
```

```python
#weight skew after treatment of train and test dataset

sns.kdeplot(num_df.weight)
sns.kdeplot(num_df_test.weight)
```

```python
print(df_ohe_test.shape,num_df_test.shape)
print(df_ohe.shape,num_df.shape)
```

```python
#updated index of test onehotencoded dataset

df_ohe_test.index=np.arange(0,len(df_ohe_test))
```

```python
# combing categorical and numerical dataset 

x_combined=pd.concat([df_ohe,num_df], axis=1)
x_combined_test=pd.concat([df_ohe_test,num_df_test],axis=1)
```

```python
# Y dataset 
y_df=train['class']
y_df=pd.DataFrame(y_df)
```

```python
# Maping of Y datset 

q_dic={"e":0,"p":1}
y_df['class']=y_df['class'].map(q_dic)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting of cross validaition set 

x_train,x_cv,y_train,y_cv= train_test_split(x_combined,y_df)
```

```python
import xgboost as xgb
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
#different classifiers

#svm classifier
clf1=svm.SVC()
#xgboost classifier
clf2=xgb.XGBClassifier(max_depth=3, n_estimators=300, learning_rate=0.05)
#KNN classifier
clf3 = KNeighborsClassifier(n_neighbors=10)
#Random forest classifier
clf4=RandomForestClassifier(n_jobs=2)
```

```python
#training differnt models

model1=clf1.fit(x_train, y_train)
model2=clf2.fit(x_train,y_train)
model3=clf3.fit(x_train,y_train)
model4=clf4.fit(x_train,y_train)
```

```python
y_cv_actual=np.array(y_cv)
```

```python
# predictions and findind accuracy on cross validation 

#SVM
print(metrics.accuracy_score(y_cv_actual,model1.predict(x_cv)))
#XGBOOST
print(metrics.accuracy_score(y_cv_actual,model2.predict(x_cv)))
#KNN
print(metrics.accuracy_score(y_cv_actual,model3.predict(x_cv)))
#Random Forest
print(metrics.accuracy_score(y_cv_actual,model4.predict(x_cv)))

```

```python
# choosing xgboost as final classifier 
#final dataframe test and train 

print(x_combined.shape)
print(y_df.shape)
print(x_combined_test.shape)
```

```python
#training on train dataset
final_model=clf2.fit(x_combined,y_df)
```

```python
#final prediction 
prediction=final_model.predict(x_combined_test)
```

```python
#predicted_datframe where 1 represent poisonous and 0 represent edible

predicted_df=pd.DataFrame(prediction)
predicted_df
```

```python
#mapping dataset
predicted_df=predicted_df.rename(columns = {0:'class'})
type_dict={0:"e",1:"p"}
predicted_df['class']=predicted_df['class'].map(type_dict)

#final result 
result_datset=pd.concat([test,predicted_df], axis=1)
result_datset
```

```python
result_datset.to_csv("DSG_Dataset_result.csv",index=False)
```
